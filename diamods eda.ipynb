### **Exploratory Data Analysis**

#### `diamonds data`

Load libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
```

Load Data

```python
df = pd.read_csv(r"C:\Users\R\Desktop\Data Analysis\R\projects\r4ds\diamonds.csv")
df.head()
```

Explore Data

```python
df.describe()
```

#### Univariate Analysis

Cut

```python
df["cut"].value_counts()
```

Bar plot

```python
sns.barplot(df["cut"].value_counts(), orient="h")
plt.xlabel("Count")
plt.ylabel("Cut")
plt.title("Bar plot of Cut");
```

Carat

```python
df["carat"].describe()
```

```python
sns.histplot(df["carat"], binwidth=0.1, kde=True)
plt.xlabel("Carat")
plt.ylabel("Frequency")
plt.title("Distribution of Carat");
```

Price

```python
df["price"].describe()
```

```python
sns.histplot(df["price"], bins = 30, kde=True)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Price");
```

Color

```python
df["color"].value_counts()
```

```python
sns.barplot(df["color"].value_counts())
plt.xlabel("Color")
plt.ylabel("Count")
plt.title("Count of Diamonds by Color");
```

```python
sns.barplot(df["clarity"].value_counts(), orient="h")
plt.ylabel("Clarity")
plt.xlabel("Count")
plt.title("Count of Diamonds by Clarity");
```

#### Bivariate Analysis

Cut and Price

```python
df.groupby("cut")["price"].mean().reset_index()
```

```python
sns.boxplot(x=df["cut"], y=df["price"])
plt.xlabel("Cut")
plt.ylabel("Price")
plt.title("Box plot of Price by Cut");
```

Carat and Price

```python
sns.scatterplot(x="carat", y="price", data=df)
plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Scatter plot of Price by Carat and Price");
```

Clarity and Price

```python
sns.violinplot(x=df["clarity"], y=df["price"])
plt.xlabel("Clarity")
plt.ylabel("Price")
plt.title("Violin plot of Price by Clarity");
```

Clarity and Depth

```python
sns.violinplot(x=df["clarity"], y=df["depth"])
plt.xlabel("Clarity")
plt.ylabel("Depth")
plt.title("Violin plot of Depth by Clarity");
```
