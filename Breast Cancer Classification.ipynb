### **Logistic Regression Model**

#### **Problem:**
Classifiy tumors as either benign or malignant

#### **`Libararies`**

```python
import warnings
warnings.filterwarnings('ignore')  

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn.datasets
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

#### **`Load Data`**

```python
df = sklearn.datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(df.data, columns=df.feature_names)
```

Add "target" column to the data

```python
df['target'] = sklearn.datasets.load_breast_cancer().target
```

#### **`Train-Test Split`**

```python
X = df.drop(columns="target", axis=1)
y = df["target"]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#### **`Model Training`**  
Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

#### **`Model Evaluation`**
Accuracy Score

```python
X_pred = model.predict(X_train)

train_data_acc = round(accuracy_score(X_pred, y_train)*100, 2)
print("Training Data Accuracy: ", train_data_acc, "%")
```

```python
X_test_pred = model.predict(X_test)

test_data_acc = round(accuracy_score(X_test_pred, y_test)*100, 2)
print("Testing Data Accuracy: ", test_data_acc, "%")
```

#### **`Building a Predicitve System`**

```python
input_data = (18.25,19.98,119.6,1040,0.09463,0.109,0.1127,
              0.074,0.1794,0.05742,0.4467,0.7732,3.18,53.91,0.004314,
              0.01382,0.02254,0.01039,0.01369,0.002179,22.88,27.66,153.2,
              1606,0.1442,0.2576,0.3784,0.1932,0.3063,0.08368)

input_data_as_numpy_array = np.asarray(input_data)

#reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
prediction
```

```python
if (prediction[0] == 0):
    print("The Breast Cancer is Malignant")
else:
    print("The Breast Cancer is Benign")
```
